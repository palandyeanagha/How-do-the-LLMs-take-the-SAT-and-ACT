```python
def get_separated_acc(data_path):
  print("Data name:", data_path)
  df = pd.read_csv(path_to_nlu_dir + data_path)

  tot_num = len(df)
  trunc_num = df['truncated'].sum()
  print("Total number:", tot_num)
  print("Number truncated:", trunc_num)
  print("Number not truncated:", tot_num - trunc_num)

  if trunc_num > 0:
    df_trunc = df[df['truncated'] == True]
    num_corr_trunc = (df_trunc['pred'] == df_trunc['correct']).sum()
    print("Truncated accuracy:", num_corr_trunc/trunc_num)
  else:
    print("Truncated accuracy: N/a")

  df_nottrunc = df[df['truncated'] == False]

  num_corr_nottrunc = (df_nottrunc['pred'] == df_nottrunc['correct']).sum()

  print("Not truncated accuracy:", num_corr_nottrunc/(tot_num - trunc_num))
  print()
```

```python
for d in data_paths:
  get_separated_acc(d)
```
